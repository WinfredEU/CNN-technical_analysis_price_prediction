```python
import pandas as pd
from datetime import datetime, timedelta
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
df=pd.read_pickle('/Users/benbilly3/Desktop/資策會專題/rawMaterialPricePrediction/RM_Price/rawMaterialPrice.pickle')
df2=df.reset_index()
#抽出月份
df2['YM']=df2['Date'].apply(lambda s:s.strftime('%Y%m'))
df2
```

# 波動率函式

```python
maxNum=df2.groupby(['stock_id','YM'])['Open','High','Low','Close','Volume'].max()
minNum=df2.groupby(['stock_id','YM'])['Open','High','Low','Close','Volume'].min()
priceFluctuation=(maxNum['High']-minNum['Low'])/maxNum['High']*100
stock_id_list=priceFluctuation.index.get_level_values(0)
stock_id_list = list(dict.fromkeys(stock_id_list))#len(stock_id_list)=15

def CFMS(stock_id,groupbyIndex):
    CF=priceFluctuation[stock_id]
    CFMS=CF.groupby([groupbyIndex]).mean()
    CFMSdf=pd.DataFrame({'CF':CFMS})
    CFMSdf=CFMSdf.reset_index()
    CFMSdf['M']=CFMSdf[groupbyIndex].apply(lambda s:s[4:])
    return CFMSdf.groupby(['M'])['CF'].mean()

CFMS('cmeCorn','YM')
```

# 波動率月份分佈繪圖

```python
#設定繪圖控制
fig, axs = plt.subplots(5,3, figsize=(30, 12), facecolor='ghostwhite', edgecolor='k')
fig.subplots_adjust(left=None, bottom=0.4, right=1.5, top=3,hspace = .3, wspace=.15)
plt.rcParams['font.size']=24
plt.rcParams['font.family']=['Arial Unicode MS']

#設定組圖變數
plotList=[[v,x] for v,x in zip(list(range(len(stock_id_list))),stock_id_list)]
axs = axs.ravel()

for i,j in plotList:
    #設定顏色
    dfa=pd.DataFrame({'pf':CFMS(j,'YM')}) 
    dfa['rank']=dfa['pf'].rank()#大小分級
    dfa['colors'] = ['indianred' if x>9 else 'royalblue' if x<4 else 'wheat' for x in dfa['rank']]
    
    axs[i].set_facecolor('snow')
    axs[i].bar(range(len(dfa['pf'])),dfa['pf'],color=dfa['colors'],linewidth=3.2,alpha=0.7)
    axs[i].set_xticks(range(len(dfa['pf'])))
    axs[i].set_xticklabels(dfa['pf'].index,fontsize='medium')
    axs[i].set_ylim(dfa['pf'].min()*0.8,dfa['pf'].max()*1.05)
    axs[i].set_xlabel('Month')
    axs[i].set_ylabel('%')
    axs[i].set_title('Price Fluctuation--'+j,fontsize='xx-large',color='darkslateblue')
    axs[i].grid(color='b', linestyle='-', linewidth=0.5)
    
```

# 月份漲跌勝率分佈繪圖

```python
df=pd.read_pickle('/Users/benbilly3/Desktop/資策會專題/rawMaterialPricePrediction/RM_Price/rawMaterialPrice.pickle')
df2=df.reset_index()
#抽出月份
df2['YM']=df2['Date'].apply(lambda s:s.strftime('%Y%m'))
df2['M']=df2['YM'].apply(lambda s:s[4:])
df2=df2.set_index('stock_id')
dataset=[]
for i in stock_id_list :
    df3=df2.loc[i]
    df3['YM']=df3['YM'].apply(lambda s:pd.to_numeric(s, errors='coerce'))
    df3['YM-2']=df3['YM'].shift(1)
    df3['newMonth']=(df3['YM']-df3['YM-2']).fillna(1)
    upAndDownS=df3[df3['newMonth']!=0]
    upAndDownS['nextMonthClose']=upAndDownS['Close'].shift(1)
    upAndDownS['upAndDownRatio']=round((upAndDownS['Close']/upAndDownS['nextMonthClose']-1)*100,2).shift(-1)
    upAndDownS=upAndDownS[~upAndDownS['upAndDownRatio'].isna()]
    upAndDownS['win']=upAndDownS['upAndDownRatio'].apply(lambda s:1 if s>0 else 0)
    dataset.append(upAndDownS)
upAndDownS=pd.concat(dataset)
upAndDownS
```

```python
t1=upAndDownS.groupby(['stock_id','M'])['upAndDownRatio'].count()
t2=upAndDownS.groupby(['stock_id','M'])['upAndDownRatio'].sum()
t3=upAndDownS.groupby(['stock_id','M'])['win'].sum()
upAndDownMean=round(t2/t1,2)
winRatio=round(t3/t1*100,2)
```

```python
#設定繪圖控制
fig, axs = plt.subplots(5,3, figsize=(30, 12), facecolor='ghostwhite', edgecolor='k')
fig.subplots_adjust(left=None, bottom=0.4, right=1.5, top=3,hspace = .3, wspace=.15)
plt.rcParams['font.size']=24
plt.rcParams['font.family']=['Arial Unicode MS']

#設定組圖變數
plotList=[[v,x] for v,x in zip(list(range(len(stock_id_list))),stock_id_list)]
axs = axs.ravel()

for i,j in plotList:
    #設定顏色
    dfa=pd.DataFrame({'pf':winRatio[j]}) 
    dfa['rank']=dfa['pf'].rank()#大小分級
    dfa['colors'] = ['indianred' if x>9 else 'royalblue' if x<4 else 'wheat' for x in dfa['rank']]
    
    axs[i].set_facecolor('snow')
    axs[i].bar(range(len(dfa['pf'])),dfa['pf'],color=dfa['colors'],linewidth=3.2,alpha=0.7)
    axs[i].set_xticks(range(len(dfa['pf'])))
    axs[i].set_xticklabels(dfa['pf'].index,fontsize='medium')
    axs[i].set_ylim(dfa['pf'].min()*0.8,dfa['pf'].max()*1.05)
    axs[i].set_xlabel('Month')
    axs[i].set_ylabel('winRatio %')
    axs[i].set_title('Price Up&Down Statistic--'+j,fontsize='xx-large',color='darkslateblue')
    axs[i].grid(color='b', linestyle='-', linewidth=0.5)
    axs[i]=axs[i].twinx()
    
    axs[i].plot(range(len(upAndDownMean[j])),upAndDownMean[j],'--',color='#FF0000',label=j,linewidth=7)
    axs[i].set_ylabel('mean %')
    axs[i].axhline(y=0,color='seagreen',linewidth=5)
```
